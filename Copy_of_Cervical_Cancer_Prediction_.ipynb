```python
input_df = df.drop(columns = 'Biopsy', axis = 1)
target_df = df['Biopsy']
```

```python
x = np.array(input_df).astype('float64')
y = np.array(target_df).astype('float64')
```

```python
x.shape, y.shape
```

```python
y.reshape(-1,1) #converting in 2D array
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
predict = lr.predict(x_test)
predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr_acc = accuracy_score(predict, y_test)
lr_acc
```

```python
y_train_pred = lr.predict(x_train)
train_accuracy = accuracy_score(y_train, y_train_pred)

# Evaluate the model on the validation data
y_val_pred = lr.predict(x_test)
val_accuracy = accuracy_score(y_test, y_val_pred)

# Compare training and validation accuracies
print(f"Training Accuracy: {train_accuracy:.2f}")
print(f"Validation Accuracy: {val_accuracy:.2f}")

# Check for overfitting
if train_accuracy > val_accuracy:
    print("Warning: Model may be overfitting.")
else:
    print("Model is not overfitting.")
```
